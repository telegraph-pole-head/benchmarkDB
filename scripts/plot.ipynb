```python
import pandas as pd
import plotly.express as px

# Read CSV data into a DataFrame
df = pd.read_csv("../data/results/benchmark1_results.csv")

# Melt the DataFrame to have Operation and Time columns
df_melted = df.melt(
    id_vars=["Container", "DataSize"],
    value_vars=["InsertTime(ms)", "DeleteTime(ms)", "AccessTime(ms)"],
    var_name="Operation",
    value_name="Time (ms)",
)

# Create a line plot
fig = px.line(
    df_melted,
    x="DataSize",
    y="Time (ms)",
    color="Container",
    line_dash="Operation",
    markers=True,
    title="Performance Comparison of Containers",
)

fig.show()
```
